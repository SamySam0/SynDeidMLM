```python
def format_xml_file(file_path):
  ''' Takes an XML file path, modifies phi tags containing newlines and saves the changes. '''
  tree = ET.parse(file_path)
  root = tree.getroot()
  modified = False

  # Extract tags
  for tag in root.findall('.//TAGS/*'):
    tag_info = {
      'start': int(tag.get('start')),
      'end': int(tag.get('end')),
      'text': tag.get('text'),
    }
    slice = root.find('TEXT').text[tag_info['start']:tag_info['end']]
    if '\n' in slice:
      # Replace the tag's text with the modified slice
      tag.set('text', slice.replace('\n', ' '))
      modified = True

  # Write the modified tree back to the file
  if modified:
    tree.write(file_path)

def format_xml_folder(folder_path):
  ''' Takes a folder path and format all XML files inside it. '''
  for filename in os.listdir(folder_path):
    if filename.endswith('.xml'):
      file_path = os.path.join(folder_path, filename)
      format_xml_file(file_path)
  return True
```

```python
# Fix newlines formatting in tags of train and test i2b2-2024-T1 datasets
format_xml_folder('drive/MyDrive/LiboMsc/data/i2b2_2024_T1_test')
format_xml_folder('drive/MyDrive/LiboMsc/data/i2b2_2024_T1_train')
```
