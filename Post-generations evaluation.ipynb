```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv(f'{PATH_TO_GENERATIONS_TO_EVALUATE}/generated_letters.csv')

# Convert each column to a list
letters_id = df['ID'].tolist()
ori_letters = df['Original letter'].tolist()
msk_letters = df['Masked letter'].tolist()
gen_letters = df['Generated letter'].tolist()
```

```python
import nltk
from collections import Counter
import string

nltk.download('stopwords')
from nltk.corpus import stopwords

# Get the list of stop words in English
stop_words = set(stopwords.words('english'))

# Function to rank top N frequent words excluding stopwords and punctuation
def top_n_words(text, N):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation using string.punctuation
    text = ''.join([char for char in text if char not in string.punctuation])
    # Split text into words
    words = text.split()
    # Filter out stop words
    words = [word for word in words if word not in stop_words]
    # Use Counter to count word frequency
    word_counts = Counter(words)
    # Get the top N most common words
    most_common_words = word_counts.most_common(N)
    return [i[0] for i in most_common_words]
```

```python
for i in [5, 10, 20, 50, 100]:
  l = []
  for ori, gen in zip(ori_letters, gen_letters):
    top_i_ori, top_i_gen = top_n_words(ori, i), top_n_words(gen, i)
    overlap_amount = 0
    for word in top_i_gen:
      if word in top_i_ori:
        overlap_amount += 1
    l.append(overlap_amount)
  print(sum(l)/len(l))
```
